# GLORYS12V1 SST Movie

Render an MP4 movie of daily sea surface temperature for the North Pacific.

```python
import sys
sys.path.insert(0, '..')

from ocean_viz import render_movie

# Path to config file
config_path = '../configs/glorys12_sst_demo.yaml'
```

## Step 1: Render movie

This will render all available daily timesteps as PNG frames, then encode to MP4.

```python
# Render MP4 movie
mp4_path = render_movie(config_path, fps=4)
print(f"Movie saved to: {mp4_path}")
```

## Step 2: Verify movie file

```python
from pathlib import Path

mp4_file = Path(mp4_path)
size_mb = mp4_file.stat().st_size / (1024 * 1024)
print(f"File size: {size_mb:.1f} MB")
print(f"File exists: {mp4_file.exists()}")
```

## Step 3: Movie information

The movie contains one frame per available daily timestep. Default frame rate is 4 fps (configurable).

```python
import subprocess

# Try to get video info with ffprobe if available
try:
    result = subprocess.run(
        ['ffprobe', '-v', 'error', '-show_entries', 'format=duration', '-of', 'default=noprint_wrappers=1:nokey=1:nokey=1', mp4_path],
        capture_output=True,
        text=True,
        timeout=5
    )
    if result.returncode == 0:
        duration = float(result.stdout.strip())
        print(f"Video duration: {duration:.1f} seconds")
        print(f"At 4 fps: {int(duration * 4)} frames")
except:
    print("ffprobe not available, cannot extract video information")
```
